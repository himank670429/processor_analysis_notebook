```python
import pandas as pd
import seaborn as sns

cpu = pd.read_csv("cpu.csv")
core_count = pd.DataFrame()
launch_date = pd.DataFrame()
TDP = pd.DataFrame()
```

```python
print(cpu.columns)
cpu
```

```python
# generation colum
core_count['generation'] = cpu['generation'].unique()

core_count['min_core'] = core_count['generation'].apply(lambda x : cpu[cpu['generation'] == x]['core_count'].min())
core_count['max_core'] = core_count['generation'].apply(lambda x : cpu[cpu['generation'] == x]['core_count'].max())
core_count['avg_core'] = core_count['generation'].apply(lambda x : cpu[cpu['generation'] == x]['core_count'].agg(lambda x: (x.min()+x.max())/2))

core_count.sort_values(by='generation', inplace=True)
core_count.reset_index()
core_count
```

```python
launch_date['year'] = cpu['launch year'].unique()

launch_date['i3'] = launch_date['year'].apply(lambda x : len(cpu[(cpu['series'] == 'i3') & (cpu['launch year'] == x)]))
launch_date['i5'] = launch_date['year'].apply(lambda x : len(cpu[(cpu['series'] == 'i5') & (cpu['launch year'] == x)]))
launch_date['i7'] = launch_date['year'].apply(lambda x : len(cpu[(cpu['series'] == 'i7') & (cpu['launch year'] == x)]))
launch_date['i9'] = launch_date['year'].apply(lambda x : len(cpu[(cpu['series'] == 'i9') & (cpu['launch year'] == x)]))

launch_date.sort_values(by = 'year', inplace=True)
launch_date.reset_index()
launch_date
```

```python
TDP['generation'] = cpu['generation'].unique()
TDP['min TDP'] = TDP['generation'].apply(lambda x: cpu[cpu['generation']==x]['TDP'].min())
TDP['max TDP'] = TDP['generation'].apply(lambda x: cpu[cpu['generation']==x]['TDP'].max())
TDP['avg TDP'] = TDP['generation'].apply(lambda x: cpu[cpu['generation']==x]['TDP'].agg(lambda x : (x.max()+x.min())/2))

TDP.sort_values(by='generation', inplace=True)
TDP.reset_index()
TDP
```

```python
cpu.to_csv('data/cpu.csv', index=False)
core_count.to_csv('data/core_count.csv', index=False)
launch_date.to_csv('data/launch_date.csv', index=False)
TDP.to_csv('data/TDP.csv', index=False)
```
