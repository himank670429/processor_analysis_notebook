```python
import seaborn as sns
import pandas as pd
```

```python
cpu = pd.read_csv('data/cpu.csv')
TDP = pd.read_csv('data/TDP.csv')
launch_date = pd.read_csv('data/launch_date.csv')
core_count = pd.read_csv('data/core_count.csv')
```

```python
# generation v/s TDP
sns.lineplot(TDP, x='generation', y='min TDP', marker='o')
sns.lineplot(TDP, x='generation', y='max TDP', marker='o')
sns.lineplot(TDP, x='generation', y='avg TDP', marker='o')
```

```python
# generation core count
sns.lineplot(core_count, x='generation', y='min_core', marker='o')
sns.lineplot(core_count, x='generation', y='max_core', marker='o')
sns.lineplot(core_count, x='generation', y='avg_core', marker='o')
```
